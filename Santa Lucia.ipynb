```python
df = pd.read_excel("Datos Demanda.xlsx")
import math
# Extraer los valores de la columna 5 (índice 4, ya que los índices comienzan en 0)
Tasa_Panzenu_1 = df["Tasa Santa Lucia Sentido 1"].tolist()
nueva_lista_Panzenu_1 = []

# Proceso
for numero in Tasa_Panzenu_1:
    valor_dividido = math.ceil(numero)
    nueva_lista_Panzenu_1.extend([valor_dividido] * 5)

# Mostrar la lista
print(nueva_lista_Panzenu_1)
len(nueva_lista_Panzenu_1)
```

```python
Tasa_Panzenu_2 = df["Tasa Santa Lucia Sentido 1"].tolist()
nueva_lista_Panzenu_2 = []

# Proceso
for numero in Tasa_Panzenu_2:
    valor_dividido = math.ceil(numero)
    nueva_lista_Panzenu_2.extend([valor_dividido] * 5)

# Mostrar la lista
print(nueva_lista_Panzenu_2)
len(nueva_lista_Panzenu_2)
```

```python
tasas= [[nueva_lista_Panzenu_1[i], nueva_lista_Panzenu_2[i]] for i in range(len(nueva_lista_Panzenu_1))]
print(tasas)
```

***Formulación y ejecución del modelo***
- El modelo es planteado en GUROBI
- El modelo es solucionado y las variables de decisión son expuestas

```python
from gurobipy import Model, GRB, quicksum
import time

# Iniciar tiempo
start_time = time.time()

# Parámetros
m = 480
n = [3, 3]
alpha = 2
t_ruta = 28
C = 45
F = 2

# Crear el modelo
model = Model("Despacho_de_Buses")

# Variables de decisión
x = model.addVars(((i, t, f) for f in range(F) for i in range(n[f]) for t in range(m)), vtype=GRB.BINARY, name="x")
b = model.addVars(((t, f) for t in range(m) for f in range(F)), lb=0, name="b")
w = model.addVars(((t, f) for t in range(m) for f in range(F)), lb=0, name="w")
delta = model.addVars(((t, f) for t in range(m) for f in range(F)), lb=0, name="delta")

# Función objetivo
model.setObjective(quicksum(w[t, f] for t in range(m) for f in range(F)), GRB.MINIMIZE)

# Restricciones
# Pasajeros abordando en cada periodo t y sentido f
for t in range(m):
    for f in range(F):
        model.addConstr(delta[t, f] == tasas[t][f] * alpha)

# En cada periodo t del sentido f solo puede ser asignado un bus i
for t in range(m):
    for f in range(F):
        model.addConstr(quicksum(x[i, t, f] for i in range(n[f])) <= 1)

# La cantidad de personas esperando en el primer intervalo
for f in range(F):
    model.addConstr(w[0, f] == delta[0, f] - b[0, f])

# Restricción de pasajeros esperando en intervalos t > 0
for t in range(1, m):
    for f in range(F):
        model.addConstr(w[t, f] == w[t - 1, f] + delta[t, f] - b[t, f])

# Capacidad de abordaje en cada intervalo
for t in range(m):
    for f in range(F):
        model.addConstr(b[t, f] <= quicksum(x[i, t, f] * C for i in range(n[f])))

# Restringir los intervalos en los que se asigna un mismo bus i
for f in range(F):
    for i in range(n[f]):
        for t in range(m):
            tiempo_min_intervalos = int((2 * t_ruta) / alpha)
            for k in range(1, tiempo_min_intervalos):
                if t + k < m:
                    model.addConstr(x[i, t, f] + x[i, t + k, f] <= 1)
            if t - m >= 0:
                model.addConstr(
                    quicksum(x[i, t - k, f] for k in range(tiempo_min_intervalos, m + 1)) >= x[i, t, f]
                )

# El último despacho debe asegurar completar la ruta
t_ruta_intervals = int(t_ruta / alpha)
t_limite = m - t_ruta_intervals
for f in range(F):
    for t in range(t_limite, m):
        for i in range(n[f]):
            model.addConstr(x[i, t, f] == 0)

# Coordinar despachos entre sentidos
for i in range(n[0]):
    for t in range(m):
        t_destino = t + int(t_ruta / alpha)
        if t_destino < m:
            model.addConstr(x[i, t_destino, 1] == x[i, t, 0])
        model.addConstr(quicksum(x[i, t, 1] for t in range(m)) == quicksum(x[i, t, 0] for t in range(m)))

# Resolver el problema
model.optimize()

# Resultados
if model.status == GRB.OPTIMAL:
    print(f"Cantidad mínima de pasajeros esperando: {model.objVal}")
    print("\nAsignaciones de servicios:")
    for f in range(F):
        for i in range(n[f]):
            for t in range(m):
                if x[i, t, f].x == 1:
                    print(f"Bus {i}, Intervalo {t}, Sentido {f}")

    print("\nPasajeros abordando por intervalo y sentido:")
    for t in range(m):
        for f in range(F):
            print(f"Intervalo {t}, Sentido {f}: {b[t, f].x}")

    print("\nPasajeros que llegaron:")
    for t in range(m):
        for f in range(F):
            print(f"Intervalo {t}, Sentido {f}: {delta[t, f].x}")

    print("\nPasajeros esperando por intervalo y sentido:")
    for t in range(m):
        for f in range(F):
            print(f"Intervalo {t}, Sentido {f}: {w[t, f].x}")

else:
    print("No se encontró una solución óptima.")

# Tiempo de ejecución
end_time = time.time()
print(f"Tiempo de CPU: {end_time - start_time:.2f} segundos")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros del modelo
m = 480  # Número de intervalos de tiempo
n = [3, 3]  # Número de buses por sentido
F = 2  # Sentidos (0 y 1)

# Crear listas para almacenar los despachos
despachos = {f: [] for f in range(F)}  # Diccionario para sentidos 0 y 1

# Extraer las soluciones del modelo Gurobi
def extraer_despachos():
    for f in range(F):
        for i in range(n[f]):
            for t in range(m):
                if x[i, t, 0].x > 0.5:  # Si el bus i es despachado en t
                    despachos[f].append((t, i))

# Llamar a la función para extraer resultados
extraer_despachos()

# Configurar la gráfica para un solo sentido (Sentido 0)
sentido = 0  # Mostrar solo el sentido 0
plt.figure(figsize=(12, 6))

# Colores únicos para cada bus
colores = plt.cm.get_cmap('tab20', n[sentido])  # Mapa de colores con 'tab20'

# Graficar los despachos por bus
tiempo_por_bus = {i: [] for i in range(n[sentido])}
for t, i in despachos[sentido]:
    tiempo_por_bus[i].append(t)

for i in range(n[sentido]):
    plt.scatter(tiempo_por_bus[i], [i] * len(tiempo_por_bus[i]), color=colores(i), label=f"Bus {i+1}", alpha=0.7)

# Configuraciones de la gráfica
plt.xlabel("Intervalos de tiempo")
plt.ylabel("Buses")
plt.title("Despachos de Buses en Intervalos de Tiempo - Sentido 1")
plt.grid(True)

# Configurar saltos en los ejes
plt.xticks(np.arange(0, m+1, 20))  # Saltos de 20 en el eje X
plt.yticks(np.arange(0, n[sentido]+1, 1))  # Saltos de 1 en el eje Y

# Mostrar la gráfica sin leyenda
plt.show()
```
