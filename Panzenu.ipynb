```python
tasas= [[nueva_lista_Panzenu_1[i], nueva_lista_Panzenu_2[i]] for i in range(len(nueva_lista_Panzenu_1))]
print(tasas)
```

***Formulación y ejecución del modelo***
- El modelo es planteado en GUROBI
- El modelo es solucionado y las variables de decisión son expuestas

```python
from gurobipy import Model, GRB, quicksum
import time

# Iniciar tiempo
start_time = time.time()

# Parámetros
m = 480
n = [16, 16]
alpha = 2
t_ruta = 30
C = 45
F = 2

# Crear el modelo
model = Model("Despacho_de_Buses")

# Variables de decisión
x = model.addVars(((i, t, f) for f in range(F) for i in range(n[f]) for t in range(m)), vtype=GRB.BINARY, name="x")
b = model.addVars(((t, f) for t in range(m) for f in range(F)), lb=0, name="b")
w = model.addVars(((t, f) for t in range(m) for f in range(F)), lb=0, name="w")
delta = model.addVars(((t, f) for t in range(m) for f in range(F)), lb=0, name="delta")

# Función objetivo
model.setObjective(quicksum(w[t, f] for t in range(m) for f in range(F)), GRB.MINIMIZE)

# Restricciones
# Pasajeros abordando en cada periodo t y sentido f
for t in range(m):
    for f in range(F):
        model.addConstr(delta[t, f] == tasas[t][f] * alpha)

# En cada periodo t del sentido f solo puede ser asignado un bus i
for t in range(m):
    for f in range(F):
        model.addConstr(quicksum(x[i, t, f] for i in range(n[f])) <= 1)

# La cantidad de personas esperando en el primer intervalo
for f in range(F):
    model.addConstr(w[0, f] == delta[0, f] - b[0, f])

# Restricción de pasajeros esperando en intervalos t > 0
for t in range(1, m):
    for f in range(F):
        model.addConstr(w[t, f] == w[t - 1, f] + delta[t, f] - b[t, f])

# Capacidad de abordaje en cada intervalo
for t in range(m):
    for f in range(F):
        model.addConstr(b[t, f] <= quicksum(x[i, t, f] * C for i in range(n[f])))

# Restringir los intervalos en los que se asigna un mismo bus i
for f in range(F):
    for i in range(n[f]):
        for t in range(m):
            tiempo_min_intervalos = int((2 * t_ruta) / alpha)
            for k in range(1, tiempo_min_intervalos):
                if t + k < m:
                    model.addConstr(x[i, t, f] + x[i, t + k, f] <= 1)
            if t - m >= 0:
                model.addConstr(
                    quicksum(x[i, t - k, f] for k in range(tiempo_min_intervalos, m + 1)) >= x[i, t, f]
                )

# El último despacho debe asegurar completar la ruta
t_ruta_intervals = int(t_ruta / alpha)
t_limite = m - t_ruta_intervals
for f in range(F):
    for t in range(t_limite, m):
        for i in range(n[f]):
            model.addConstr(x[i, t, f] == 0)

# Coordinar despachos entre sentidos
for i in range(n[0]):
    for t in range(m):
        t_destino = t + int(t_ruta / alpha)
        if t_destino < m:
            model.addConstr(x[i, t_destino, 1] == x[i, t, 0])
        model.addConstr(quicksum(x[i, t, 1] for t in range(m)) == quicksum(x[i, t, 0] for t in range(m)))

# Resolver el problema
model.optimize()

# Resultados
if model.status == GRB.OPTIMAL:
    print(f"Cantidad mínima de pasajeros esperando: {model.objVal}")
    print("\nAsignaciones de servicios:")
    for f in range(F):
        for i in range(n[f]):
            for t in range(m):
                if x[i, t, f].x == 1:
                    print(f"Bus {i}, Intervalo {t}, Sentido {f}")

    print("\nPasajeros abordando por intervalo y sentido:")
    for t in range(m):
        for f in range(F):
            print(f"Intervalo {t}, Sentido {f}: {b[t, f].x}")

    print("\nPasajeros que llegaron:")
    for t in range(m):
        for f in range(F):
            print(f"Intervalo {t}, Sentido {f}: {delta[t, f].x}")

    print("\nPasajeros esperando por intervalo y sentido:")
    for t in range(m):
        for f in range(F):
            print(f"Intervalo {t}, Sentido {f}: {w[t, f].x}")

else:
    print("No se encontró una solución óptima.")

# Tiempo de ejecución
end_time = time.time()
print(f"Tiempo de CPU: {end_time - start_time:.2f} segundos")
```

***Mapa de la ruta***

```python
import osmnx as ox
import folium
import networkx as nx
# Coordenadas de la ruta
coordinates = [
    (8.801581, -75.857790),
    (8.805038, -75.858917),
    (8.806936, -75.854754),
    (8.805472, -75.850098),
    (8.804921, -75.849315),
    (8.763848, -75.878683),
    (8.751655, -75.889127),
    (8.750749, -75.890228),
    (8.748447, -75.891214),
    (8.745657, -75.893414),
    (8.744370, -75.893225),
    (8.743159, -75.893062),
    (8.734959, -75.893059),
    (8.732497, -75.893361),
    (8.725182, -75.893043),
    (8.726035, -75.892490),
    (8.729474, -75.890396),
    (8.728205, -75.888103),
    (8.728848, -75.886387),
    (8.726761, -75.887024),
    (8.725639, -75.884588),
]

# Descargar el grafo de Montería
place_name = "Montería, Córdoba, Colombia"
graph = ox.graph_from_place(place_name, network_type='drive')

# Obtener los nodos más cercanos a las coordenadas
nodes = []
for i, coord in enumerate(coordinates):
    try:
        node = ox.distance.nearest_nodes(graph, X=coord[1], Y=coord[0])
        nodes.append(node)
    except Exception as e:
        print(f"Error al encontrar el nodo más cercano para {coord}: {e}")

# Calcular la ruta que pasa por todos los nodos intermedios
full_route = []
for i in range(len(nodes) - 1):
    try:
        # Calcular ruta más corta entre nodos consecutivos
        path = nx.shortest_path(graph, nodes[i], nodes[i + 1], weight='length')
        full_route.extend(path[:-1])  # Evitar duplicar nodos
    except Exception as e:
        print(f"Error al calcular la ruta entre nodos {nodes[i]} y {nodes[i + 1]}: {e}")

# Añadir el último nodo a la ruta
full_route.append(nodes[-1])

# Crear el mapa centrado en la primera coordenada
ma = folium.Map(location=coordinates[0], zoom_start=14)

# Añadir los marcadores de las coordenadas
for coord in coordinates:
    folium.Marker(location=coord).add_to(ma)

# Obtener las coordenadas de la ruta completa
route_coords = [(graph.nodes[node]['y'], graph.nodes[node]['x']) for node in full_route]

# Trazar la ruta sobre el mapa
folium.PolyLine(route_coords, color="blue", weight=2.5, opacity=1).add_to(ma)

# Guardar el mapa como archivo HTML
ma
```

```python
total_distance = 0
for i in range(len(full_route) - 1):
    try:
        edge_data = graph.get_edge_data(full_route[i], full_route[i + 1])
        # Si hay múltiples aristas entre los nodos, selecciona la primera
        length = edge_data[0]['length'] if edge_data else 0
        total_distance += length
    except Exception as e:
        print(f"Error al obtener la longitud entre nodos {full_route[i]} y {full_route[i + 1]}: {e}")

# Imprimir la distancia total
print(f"La distancia total de la ruta es: {total_distance/1000:.2f} kilometros")
```
